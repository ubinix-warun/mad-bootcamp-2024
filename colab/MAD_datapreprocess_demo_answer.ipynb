```python
def pipeline_label(demo, txn, cohort):
    print('start build labels')
    since = cohort
    until = cohort + relativedelta(months=1) - relativedelta(days=1)
    print(f"{since}-----{until}")

    demo = demo[demo['pos_dt']==cohort][['user_id']]
    txn_data = txn[txn["pos_dt"].between(since, until)]

    cnt_txn = txn_data.groupby("user_id").agg(txn_cnt=("txn_amt", 'count'))
    label = demo.merge(cnt_txn, on='user_id', how='left').fillna(0)
    label['is_churn'] = (label['txn_cnt']==0)*1
    label = label[['user_id', 'is_churn']]
    label['cohort'] = cohort
    return label
```

```python
def filter_txn_by_cohort(txn, cohort, duration=6):
    since = cohort - relativedelta(months=duration)
    until = cohort - relativedelta(days=1)
    print(f"{since}-----{until}")
    txn_data = txn[txn["pos_dt"].between(since, until)]
    return txn_data
```

```python
filter_txn_by_cohort(txn=cc_txn, cohort=datetime(2024, 2, 1), duration=6)
```

```python


def agg_stats(txn, cohort, duration):
    print('calculate stats')
    txn_m = filter_txn_by_cohort(txn=txn, cohort=cohort, duration=duration)
    feats = txn_m.groupby("user_id").agg(
        txn_cnt=("txn_amt", 'count'),
        sum_amt=("txn_amt", 'sum'),
        max_amt=("txn_amt", 'max'),
        min_amt=("txn_amt", 'min'),
    ).reset_index()

    feats = feats.rename(columns={"txn_cnt": f"txn_cnt_l{duration}m",
                                  "sum_amt": f"sum_amt_l{duration}m",
                                  "max_amt": f"max_amt_l{duration}m",
                                  "min_amt": f"min_amt_l{duration}m",}
                        )

    feats[f"txn_cnt_per_month_l{duration}m"] = feats[f"txn_cnt_l{duration}m"]/duration
    feats[f"sum_amt_per_month_l{duration}m"] = feats[f"sum_amt_l{duration}m"]/duration
    return feats


def agg_stats_in_category(txn, cohort, duration, category):
    print(f'calculate stats in {category}')
    txn_m = filter_txn_by_cohort(txn=txn, cohort=cohort, duration=duration)
    txn_m = txn_m[txn_m['category']==category]
    feats = txn_m.groupby("user_id").agg(
        txn_cnt=("txn_amt", 'count'),
        sum_amt=("txn_amt", 'sum'),
    ).reset_index()

    feats = feats.rename(columns={"txn_cnt": f"txn_cnt_{category}_l{duration}m",
                                  "sum_amt": f"sum_amt_{category}_l{duration}m"}
                        )

    return feats
```

```python
def pipeline_features(demo, txn, cohort):
    print('start build features')
    demo = demo[demo['pos_dt']==cohort]
    demo = demo[['user_id', 'income', 'age', 'tenure', 'province', 'segment']]

    txn_agg_1 = agg_stats(txn=txn, cohort=cohort, duration=1)
    txn_agg_3 = agg_stats(txn=txn, cohort=cohort, duration=3)
    txn_agg_6 = agg_stats(txn=txn, cohort=cohort, duration=6)
    txn_gasoline_agg_3 = agg_stats_in_category(txn=txn, cohort=cohort, duration=3, category='gasoline')

    features = (
        demo.merge(txn_agg_1, on="user_id", how="left")
        .merge(txn_agg_3, on="user_id", how="left")
        .merge(txn_agg_6, on="user_id", how="left")
        .merge(txn_gasoline_agg_3, on="user_id", how="left")
    )


    features['txn_cnt_trend'] = features['txn_cnt_l1m']/features['txn_cnt_per_month_l6m']
    features['sum_amt_trend'] = features['sum_amt_l1m']/features['sum_amt_per_month_l6m']
    features['cohort'] = cohort
    features = features[
        [
            "user_id",
            "cohort",
            'age', 'segment', 'tenure', 'income',

            "txn_cnt_l1m",
            "txn_cnt_l3m",
            "txn_cnt_l6m",

            "sum_amt_l1m",
            "sum_amt_l3m",
            "sum_amt_l6m",


            "min_amt_l3m",
            "max_amt_l3m",
            "min_amt_l6m",
            "max_amt_l6m",

            "txn_cnt_per_month_l6m",
            "sum_amt_per_month_l6m",

            "txn_cnt_gasoline_l3m",
            "sum_amt_gasoline_l3m",

            "txn_cnt_trend",
            "sum_amt_trend",
        ]
    ]

    return features
```

```python
def gen_Xy(demo, txn, cohort):
    X = pipeline_features(demo, txn, cohort=cohort)
    y = pipeline_label(demo, txn, cohort=cohort)
    Xy = X.merge(y, on=['user_id', 'cohort'])
    return Xy
```
